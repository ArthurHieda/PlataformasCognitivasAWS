```python
import os
import boto3
from PIL import Image

# Configuração das credenciais da AWS
aws_access_key_id = 'SEU_ACCESS_KEY_ID' #TROCAR NÃO ESQUECE VIU???
aws_secret_access_key = 'SEU_SECRET_ACCESS_KEY' #TROCAR NÃO ESQUECE VIU???
aws_bucket_name = 'NOME_DO_BUCKET' #TROCAR NÃO ESQUECE VIU???

# Inicializa o cliente do S3
s3 = boto3.client('s3', aws_access_key_id=aws_access_key_id, aws_secret_access_key=aws_secret_access_key)

# Solicita o caminho da pasta ao usuário
folder_path = input("Digite o caminho da pasta com as fotos: ")

# Verifica se o caminho existe
if not os.path.exists(folder_path):
    print("O caminho da pasta não existe.")
else:
    # Lista os arquivos na pasta
    file_list = os.listdir(folder_path)

    # Itera pelos arquivos na pasta
    for file_name in file_list:
        if file_name.lower().endswith('.jpg'):
            # Cria o caminho completo do arquivo
            file_path = os.path.join(folder_path, file_name)

            # Abre a imagem usando o Pillow
            image = Image.open(file_path)

            # Gera um nome único para o objeto no S3
            s3_object_key = f'fotos/{file_name}'

            # Envia a imagem para o S3
            s3.upload_file(file_path, aws_bucket_name, s3_object_key)

            print(f"{file_name} enviado com sucesso para o S3.")

print("Concluído.")
```
